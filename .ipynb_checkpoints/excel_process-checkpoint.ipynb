```python
import pandas as pd
```

```python
excel_file = "./data/HetY1H Pilot TF coordinates.xlsx"
```

```python
df = pd.read_excel(excel_file, engine='openpyxl')#, index_col=None, header=None) 
```

```python
df
```

```python
def strip_string_col(x):
    try :
        res = x.split('-')
        return int(res[1][1:])
        #print("except")
    except:
        return -1
```

```python
def strip_string_row(x):
    try : 
        res = x.split('-')
        return res[1][0]
    except:
        #print("except")
        return "null"
```

```python
def strip_string_plate(x):
    try : 
        res = x.split('-')
        return int(res[0])
    except:
        #print("except")
        return -1
```

```python
def strip_string_row_col(x):
    try : 
        res = x.split('-')
        return res[1]
    except:
        #print("except")
        return "null"
```

```python
df.insert(1, "Coordinate_col", list(df["Coordinate"].apply(lambda x : strip_string_col(x))), True) 
```

```python
df.insert(2, "Coordinate_row", list(df["Coordinate"].apply(lambda x : strip_string_row(x))), True) 
```

```python
df.insert(3, "Coordinate_plate", list(df["Coordinate"].apply(lambda x : strip_string_plate(x))), True) 
```

```python
df.insert(4, "Coordinate_row_col", list(df["Coordinate"].apply(lambda x : strip_string_row_col(x))), True)
```

```python
df.sort_values(by=['Coordinate_col'])
```

```python
df = df[df.Coordinate_col != -1]
```

```python
df
```

```python
df.sort_values(by=['Coordinate_row', 'Coordinate_col', 'Coordinate_plate'], ascending=[True, True, True]) 
```
